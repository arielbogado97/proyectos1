OBJETIVO:
Crear una red Neuronal simple, que permita predecir y convertir  unidades de temperaturas, de grados Celsius a Farenheint.

1.TensorFlow.

---


2.Numpy.

---


3.Keras.

---


4.matplotlib.  










```python
import tensorflow as tf
import numpy as np
```

```python
celsius=np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit=np.array([-40,14,32,46,59,72,100], dtype=float)
```

```python
capa = tf.keras.layers.Dense(units=1,input_shape=[1])
modelo = tf.keras.Sequential ([capa])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("Entrenar red....")
historial = modelo.fit(celsius,fahrenheit,epochs=1000, verbose=False)
print("Listo para hacer predicciones")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("#Época")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])

```

```python
print("Predicción")
resultado=modelo.predict([90.0])
print("El resultado es"+ str(resultado)+"Farenheit")
```
